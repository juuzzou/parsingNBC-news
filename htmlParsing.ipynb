```python
urls = []

for link in soup.find_all('a'):
  if '/news' in link.get('href'):
    if not 'newsletters' in link.get('href'):
      if not '/weather' in link.get('href'):
        if not '/nbc-news-now-live' in link.get('href'):
          if 'https://' in link.get('href'):
            urls.append(link.get('href'))
urls
```

```python
def getNews(url0):
  """
  Here we are parsing every data we need, such as an author's name, titile of an article and a date of publishing, a link.
  """
  page0 = requests.get(url0)
  soup0 = BeautifulSoup(page0.text, 'html')

  title = soup0.find('meta', {'property': 'og:title'}).get('content')
  author = soup0.find('meta', {'name': 'branch:deeplink:authorName1'}).get('content')
  date = soup0.find('meta', {'property': 'article:published_time'}).get('content')[:10]

  return url0, author, title, date
```

```python
from time import sleep
```

```python
urls
```

```python
news = []
for link in urls:
  print(link)
  res = getNews(link)
  news.append(res)

  sleep(3)
news
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(news)
```
